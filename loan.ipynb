## Filling missing values with mode,mean

```python
loan['Gender']=loan['Gender'].fillna('Male')
loan['Married']=loan['Married'].fillna('Yes')
loan['Dependents']=loan['Dependents'].fillna('0')
loan['Self_Employed']=loan['Self_Employed'].fillna('No')
loan['LoanAmount']=loan['LoanAmount'].fillna(146.41)
loan['Loan_Amount_Term']=loan['Loan_Amount_Term'].fillna(360.0)
loan['Credit_History']=loan['Credit_History'].fillna(1.0)
```

```python
loan_test['Gender']=loan_test['Gender'].fillna('Male')
loan_test['Dependents']=loan_test['Dependents'].fillna('0')
loan_test['Self_Employed']=loan_test['Self_Employed'].fillna('No')
loan_test['LoanAmount']=loan_test['LoanAmount'].fillna(146.41)
loan_test['Loan_Amount_Term']=loan_test['Loan_Amount_Term'].fillna(360.0)
loan_test['Credit_History']=loan_test['Credit_History'].fillna(1.0)
    
```

# Adding New Feature

```python
loan['ApplicantIncome_CoapplicantIncome']=loan['ApplicantIncome']+loan['CoapplicantIncome']
loan['App+Coapp-LoanAmt']=loan['ApplicantIncome_CoapplicantIncome']-loan['LoanAmount']
loan_test['ApplicantIncome_CoapplicantIncome']=loan_test['ApplicantIncome']+loan_test['CoapplicantIncome']
loan_test['App+Coapp-LoanAmt']=loan_test['ApplicantIncome_CoapplicantIncome']-loan_test['LoanAmount']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVC
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score
from sklearn.grid_search  import GridSearchCV
```

```python
predictors=['Credit_History','Education','Married','Dependents','App+Coapp-LoanAmt']
kf = KFold(len(loan), 3, shuffle=False, random_state=8)
clf=LogisticRegression()
accuracies = cross_val_score(clf,loan[predictors],loan['Loan_Status'], scoring="accuracy", cv=kf)
accuracies.mean()
```
